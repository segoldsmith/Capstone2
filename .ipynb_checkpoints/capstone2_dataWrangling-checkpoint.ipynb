```python
#data collection: fi stands for flight information
fi = pd.read_csv("../raw_data/2018.csv")

#get overview of the data
fi.info(show_counts=True)
```

```python
#drop irrelevant columns (8, 9, 10, 11, 17, 18, 19, 20, 27)
fi_drop = fi.drop(columns=['TAXI_OUT', 'WHEELS_OFF', 'WHEELS_ON', 'TAXI_IN', 'DIVERTED', 'CRS_ELAPSED_TIME', 'CRS_ELAPSED_TIME',\
                          'ACTUAL_ELAPSED_TIME', 'AIR_TIME', 'Unnamed: 27'])
fi_drop.columns
```

```python
#rename columns
fi_drop_rename = fi_drop.rename(columns = {'FL_DATE':'Date', 'OP_CARRIER':"Airline", 'OP_CARRIER_FL_NUM':'Flight_Number',\
                                          'CRS_DEP_TIME':'Planned_Depart_Time','DEP_TIME':'Actual_Depart_Time',\
                                           'Dep_delay':'Depart_Delay', 'CRS_ARR_TIME':'Planned_Arrival_Time',\
                                           'ARR_TIME':'Actual_Arrival_Time', 'ARR_DELAY':'Arrival_Delay',})
fi_drop_rename.columns = fi_drop_rename.columns.str.capitalize()
fi_drop_rename.columns
```

```python
#change date column to datetime
fi_drop_rename['Date'] = pd.to_datetime(fi_drop_rename['Date']).dt.date
fi_drop_rename.head()
```

```python
#change time columns (5, 6, 8, 9) to datetime
fi_drop_rename['Planned_depart_time'] = fi_drop_rename['Planned_depart_time'].astype(str).str.zfill(4)
fi_drop_rename['Planned_depart_time'] = pd.to_datetime(fi_drop_rename['Planned_depart_time'], format= "%H%M",\
                                                       errors='coerce').dt.time

fi_drop_rename['Actual_depart_time'] = fi_drop_rename['Actual_depart_time'].astype(str)
fi_drop_rename['Actual_depart_time'] = fi_drop_rename['Actual_depart_time'].str.split('.').str[0]
fi_drop_rename['Actual_depart_time'] = fi_drop_rename['Actual_depart_time'].str.zfill(4)
fi_drop_rename['Actual_depart_time'] = pd.to_datetime(fi_drop_rename['Actual_depart_time'], format= "%H%M",\
                                                      errors='coerce').dt.time

fi_drop_rename['Planned_arrival_time'] = fi_drop_rename['Planned_arrival_time'].astype(str).str.zfill(4)
fi_drop_rename['Planned_arrival_time'] = pd.to_datetime(fi_drop_rename['Planned_arrival_time'], format= "%H%M",\
                                                        errors='coerce').dt.time

fi_drop_rename['Actual_arrival_time'] = fi_drop_rename['Actual_arrival_time'].astype(str)
fi_drop_rename['Actual_arrival_time'] = fi_drop_rename['Actual_arrival_time'].str.split('.').str[0]
fi_drop_rename['Actual_arrival_time'] = fi_drop_rename['Actual_arrival_time'].str.zfill(4)
fi_drop_rename['Actual_arrival_time'] = pd.to_datetime(fi_drop_rename['Actual_arrival_time'], format= "%H%M",\
                                                       errors='coerce').dt.time

```
